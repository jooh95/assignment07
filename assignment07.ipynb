```python
import numpy as np
import matplotlib.pyplot as plt
```

# Define the function.

```python
num     = 1001
std     = 5 

def fun(x):

	f = np.abs(x) * np.sin(x)
	return f
```

# Define polynomial least square function.

```python
def least_square(x, y, p):

    n_x = np.zeros((p+1, num))
    for i in range(0,p+1):
        n_x[i] = np.array([x ** i])
    n_x = n_x.T

    s_inverse = np.dot(np.linalg.inv(np.dot(n_x.T, n_x)), n_x.T)
    
    return np.dot(s_inverse, y)
```

# Set clean and noisy Data.

```python
n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-10,10,num)
y1      = fun(x)
y2      = y1 + nn * std
```

# Plot clean data.

```python
plt.title('Clean Data')
plt.plot(x, y1, 'r.')
plt.show()
```

# Plot noisy data.

```python
plt.title('Noisy Data')
plt.plot(x, y2, 'g.')
plt.show()
```

# Plot least square approximated data and error value.

```python
for i in range(0, 10):
    plt.title('Least Square Approximation by p = %d' %i)
    p = least_square(x, y2, i)
    
    y3 = 0
    for j in range(0, i+1):
        y3 += p[j] * x ** j
        
    plt.plot(x, y3, 'b.')
    plt.show()

    error = y1 - y3
    error = error ** 2
    
    plt.title('Error when p = %d' %i)
    plt.plot(x, error, 'k.')
    plt.show()  
```
